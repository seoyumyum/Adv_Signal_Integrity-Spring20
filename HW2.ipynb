```python
import numpy as np
from numpy.linalg import*
import matplotlib.pyplot as plt
from numpy import r_, exp, cos, sin, pi, zeros, ones, hanning, sqrt, log, floor, reshape, mean
from scipy import signal
from numpy.fft import fft
import math
#%matplotlib inline
```

```python
#dimensions
l=1 # m
a = 0.001 # m
ep_0 = 8.854e-12
N = 100 # Number of sections
A = np.zeros((N,N))  # Constant Matrix
V_total=np.ones((N,1))
def Cof_k2i(k,i,N):
    if k==i:
        return (a/ep_0)*np.log(l/(N*a))
    elif k>i:
        return (a/(2*ep_0))*np.log((k-i+0.5)/(k-i-0.5))
    elif k<i:
        return (a/(2*ep_0))*np.log((k-i-0.5)/(k-i+0.5))

for i in range(N):
    for k in range(N):
        A[i,k]=(Cof_k2i(k+1,i+1,N))

Charge_den = np.matmul(np.linalg.inv(A),V_total)
#print(Charge_den)
Charge = 2*np.pi*a*(l/N)*Charge_den
plt.plot(Charge_den)
plt.xlabel(r'Section #')
plt.ylabel(r'Charge [C]')
#plt.xlim([-2,50])
plt.grid()
```
